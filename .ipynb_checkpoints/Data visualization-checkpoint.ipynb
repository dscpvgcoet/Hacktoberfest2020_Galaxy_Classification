```python
import numpy as np
import matplotlib.pyplot as plt
data = np.load('galaxy_catalogue.npy')

figure = plt.figure()
plt.plot(data['ecc'],data['class'],'ro',alpha=0.1)
figure.suptitle('ecc by Galaxy type', fontsize=12)
```

```python
figure = plt.figure()
plt.plot(data['u-g'],data['class'],'ro',alpha=0.1)
figure.suptitle('u-g by Galaxy type', fontsize=12)
```

```python
figure = plt.figure()
plt.plot(data['g-r'],data['class'],'ro',alpha=0.1)
figure.suptitle('g-r by Galaxy type', fontsize=12)
```

```python
figure = plt.figure()
plt.plot(data['r-i'],data['class'],'ro',alpha=0.1)
figure.suptitle('r-i by Galaxy type', fontsize=12)
```

```python
figure = plt.figure()
plt.plot(data['i-z'],data['class'],'ro',alpha=0.1)
figure.suptitle('i-z by Galaxy type', fontsize=12)
```

```python
figure = plt.figure()
plt.plot((data['u-g']+data['g-r']+data['r-i']+data['i-z'])*10**10,data['class'],'ro',alpha=0.1)
figure.suptitle('All Color Indices by Galaxy type', fontsize=12)
```
